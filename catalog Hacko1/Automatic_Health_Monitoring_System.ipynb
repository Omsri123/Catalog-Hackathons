```python
class Patient:
    def __init__(self, name, age, patient_id):
        self.name = name
        self.age = age
        self.patient_id = patient_id

    def __str__(self):
        return f"Name: {self.name}, Age: {self.age}, Patient ID: {self.patient_id}"


class DiagnosisReport:
    def __init__(self, report_id, patient_id, diagnosis_details):
        self.report_id = report_id
        self.patient_id = patient_id
        self.diagnosis_details = diagnosis_details

    def __str__(self):
        return f"Report ID: {self.report_id}, Diagnosis: {self.diagnosis_details}"


class HealthMonitoringSystem:
    patients = {}
    reports = {}

    @staticmethod
    def main():
        while True:
            print("Welcome to the Automatic Health Monitoring System")
            print("1. Register Patient")
            print("2. Add Diagnosis Report")
            print("3. View Patient Reports")
            print("4. Exit")

            choice = input()

            if choice == '1':
                HealthMonitoringSystem.register_patient()
            elif choice == '2':
                HealthMonitoringSystem.add_diagnosis_report()
            elif choice == '3':
                HealthMonitoringSystem.view_patient_reports()
            elif choice == '4':
                print("Exiting the system. Goodbye!")
                break
            else:
                print("Invalid choice, please try again.")

    @staticmethod
    def register_patient():
        name = input("Enter Patient Name: ")
        age = int(input("Enter Patient Age: "))
        patient_id = input("Enter Patient ID: ")

        patient = Patient(name, age, patient_id)
        HealthMonitoringSystem.patients[patient_id] = patient
        print("Patient registered successfully!")

    @staticmethod
    def add_diagnosis_report():
        patient_id = input("Enter Patient ID: ")

        if patient_id not in HealthMonitoringSystem.patients:
            print("Patient not found. Please register the patient first.")
            return

        report_id = input("Enter Report ID: ")
        diagnosis_details = input("Enter Diagnosis Details: ")

        report = DiagnosisReport(report_id, patient_id, diagnosis_details)
        HealthMonitoringSystem.reports[report_id] = report
        print("Diagnosis report added successfully!")

    @staticmethod
    def view_patient_reports():
        patient_id = input("Enter Patient ID: ")

        if patient_id not in HealthMonitoringSystem.patients:
            print("Patient not found.")
            return

        print(f"Diagnosis Reports for Patient ID: {patient_id}")
        for report in HealthMonitoringSystem.reports.values():
            if report.patient_id == patient_id:
                print(report)


HealthMonitoringSystem.main()
```
